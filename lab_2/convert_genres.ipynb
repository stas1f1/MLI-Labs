```python
import pandas as pd
import numpy as np
import os
```

```python
genres_list = ['Action','Adventure','Animation','Biography','Comedy','Crime','Documentary','Drama','Family','Fantasy','Film-Noir','History','Horror','Music','Musical','Mystery','Romance','Sci-Fi','Short Film','Sport','Superhero','Thriller','War','Western']
```

```python
genres_dict = dict()

for i in range(len(genres_list)):
  genres_dict[genres_list[i]] = i
```

```python
film_data_df = pd.read_csv("/content/drive/MyDrive/PAPI_NASONOV/lab2/movie_info_final.csv")
```

```python
def convert_genres_to_categorical(genres):
  return [1 if genre in genres else 0 for genre in genres_list]
```

```python
genres_col = film_data_df.genres.tolist()
for i in error:
  genres_col[i] = "['Comedy', 'Comedy']"
genres_col = [[ genre for genre in x[2:-2].split("', '") if genre in genres_list] for x in genres_col]
```

```python
genres_res = [genres_col[0][0]]
```

```python
for i in range(1,len(genres_col)-1):
  intersec = [value for value in genres_col[i] if value in genres_col[i-1] and value in genres_col[i+1]]
  val = genres_col[i][0]
  if len(intersec) > 0:
    val = intersec[0]
  genres_res.append(val)
```
